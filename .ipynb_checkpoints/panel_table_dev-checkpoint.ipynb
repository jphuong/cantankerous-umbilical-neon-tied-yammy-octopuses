### What are the trends in drug overdose death rates over time? Highlight any significant changes or patterns in the death rates over the years. Break down the trends by 

 - Drug overdose type
 - Sex
 - Age
 - Race

```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import os
import scipy
import numpy as np

%matplotlib inline
```

```python
# test read source data
source_file = os.path.join(os.getcwd(),'Drug_overdose_death_rates__by_drug_type__sex__age__race__and_Hispanic_origin__United_States_20240518.csv')
opioid_df = pd.read_csv(source_file)

# field cleaning
opioid_df.STUB_NAME = opioid_df.STUB_NAME.replace({'Sex and race (single race)':'Sex and race',
                                                 'Sex and race and Hispanic origin (single race)':'Sex and race and Hispanic origin'})

opioid_df.head(50)
```

```python
opioid_df.UNIT.unique()
```

```python
opioid_df.PANEL.unique()
```

```python
pd.pivot_table(data=opioid_df.query('UNIT=="Deaths per 100,000 resident population, age-adjusted"'),
               index=['YEAR'],
               columns=['STUB_NAME','AGE','PANEL'],
               values=['ESTIMATE'])
```

```python
pd.pivot_table(data=opioid_df.query('UNIT=="Deaths per 100,000 resident population, crude"'),
               index=['YEAR'],
               columns=['STUB_NAME','AGE','PANEL'],
               values=['ESTIMATE'])
```
